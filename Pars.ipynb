```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time, re, pandas as pd

options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome("D:/Trash/chromedriver_win32/chromedriver.exe", options=options)
```

```python
alfav = [u'Ц', u'Ч', u'Ш', u'Щ', u'Ы', u'Э', u'Ю', u'Я']

#alfav = [u'А', u'Б', u'В', u'Г', u'Д', u'Е', u'Ж', u'З', u'И', u'Й', u'К', u'Л', u'М', u'Н', u'О',
#         u'П', u'Р', u'С', u'Т', u'У', u'Ф', u'Х', u'Ц', u'Ч', u'Ш', u'Щ', u'Ы', u'Э', u'Ю', u'Я']

def stranich_v_bukve(character):
    
    my_url = 'https://search.rsl.ru/ru/catalog#p=1&d=xdis&ltr=' + character + '&st=author'
    driver.get(my_url)
    time.sleep(1)
    
    html_source = driver.page_source
    soup = BeautifulSoup(html_source,'html.parser')
    results = soup.findAll('a',{'class':'pagination-link'})
    
    how_many_pages = int(results[-1]['data-page']) #число страниц в букве
    
    return how_many_pages

def nazvania_disser(character, how_many_pages):

    d = [None] * how_many_pages
    fl_name = character + u'_disser_names.csv'
    
    for i in range(1, how_many_pages):
        my_url = 'https://search.rsl.ru/ru/catalog#p=' + str(i) + '&d=xdis&ltr=' + character + '&st=author'
        driver.get(my_url)
        time.sleep(1)
        
        html_source = driver.page_source
        soup = BeautifulSoup(html_source,'html.parser')
        results = soup.findAll('span',{'class':'js-item-maininfo'})
        
        for x in results:
            text_body = re.split(':', x.text)[0]
            d.append(text_body)

        print u'Буква: ', character, u'\tСтраница: ', i, u'\tиз ', how_many_pages
            
    df_ = pd.DataFrame(d)
    df_.to_csv(fl_name, encoding='utf-8', index = False)
```

```python
for i in alfav:
    print i, '\t', stranich_v_bukve(i)
    nazvania_disser(i, stranich_v_bukve(i))
    #nazvania_disser(i, 3)
```
